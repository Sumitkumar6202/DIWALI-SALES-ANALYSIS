```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib  inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv',encoding="unicode_escape")
df
```

```python
df.info()
```

```python
df.drop(['Status', 'unnamed1'],axis=1,inplace=True)
```

```python
df
```

```python
df.shape
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

df.rename(columns={'Marital_Status':'Shadi'})

```python
df.describe()
```

Explorlatory Data Analysis

```python
ax = sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
    
    
                  
```

```python
sales_gen=df.groupby(['Gender'],as_index= False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sales_gen
```

```python
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of women are greater then men

```python
ax = sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index= False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sales_age
```

```python
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

from above age  we can see that  most of the buyers are of age group between 26-35

```python
#TOP10 order states
sales_state=df.groupby(['State'],as_index= False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sales_state
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
sales_state=df.groupby(['State'],as_index= False)['Amount'].sum().sort_values(by='Amount',ascending=True).head(10)
sales_state
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=sales_state)
```

from above two graphs we can see that most order is from uttar pradesh and most amount is spend by himachal pradeh

Marital status

```python
ax = sns.countplot(data=df,x='Marital_Status')

for bars in ax.containers:
    sns.set(rc={'figure.figsize':(5,5)})
    ax.bar_label(bars)
```

```python
sales_Marital_Status=df.groupby(['Marital_Status','Gender'],as_index= False)['Amount'].sum().sort_values(by='Amount',ascending=True).head(10)
sales_Marital_Status
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=sales_Marital_Status)
```

from above graph we can see that  most of the buyers are married(women)

*OCCUPATION*

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data=df,x='Occupation')

for bars in ax.containers:
    
    ax.bar_label(bars)
```

```python
sales_occupation=df.groupby(['Occupation'],as_index= False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='Occupation',y='Amount',data= sales_occupation)
```

*From above Graphs we can see that most of the worker are working in IT Sector and Healthcare*
